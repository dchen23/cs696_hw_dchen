```python
def df_mean(filename):
    import pandas as pd
    
    df = pd.read_csv(filename, sep=';')
    d = df.iloc[:,2:]
    return d.mean()
```

```python
def df_median(filename):
    import pandas as pd
    
    df = pd.read_csv(filename, sep=';')
    d = df.iloc[:,2:]
    return d.median()
```

```python
def df_std(filename):
    import pandas as pd
    
    df = pd.read_csv(filename, sep=';')
    d = df.iloc[:,2:]
    return d.std()
```
