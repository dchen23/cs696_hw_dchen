## K-means way to approach

For here, we extract our label from original dataset, then convert them into **0** and **1**.

```python
import pandas as pd
    
df = pd.read_csv("feature_selection.csv", sep=',')
y = df.iloc[:,2].astype('str')

for i in range(len(y)):
    if (y[i]  == 'typ'):
        y[i] = 1
    else:
        y[i] = 0
y.head()
```

Then I tried k-means on the frist channel of group A, then get the predict array and count out the errors.

```python
index = 3
X = df.iloc[:,index: index+5]
print(X.head())
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 2)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
kmeans.inertia_
```

Since I converted the label randomly, so the label only make a different for the data. So either very low errors and very high errors mean that the feature make the data spereate close to the label classification.

```python
error = np.sum(np.abs(y_kmeans - y))
error
```

I try each channel using k-means to count out the errors and select all low errors channel and high errors channel. 

```python
for i in range(90):
    index = 3 + i*5
    X = df.iloc[:,index: index+5]
    from sklearn.cluster import KMeans
    kmeans = KMeans(n_clusters = 2)
    kmeans.fit(X)
    y_kmeans = kmeans.predict(X)
    error = np.sum(np.abs(y_kmeans - y))
    if (error < 10 or error > 800):
        print(str(i) + ': ' + str(np.sum(np.abs(y_kmeans - y))))
```

According to the output, the significant channel are (A.c9, B.c1, B.c3, C.c1, E.c2, F.c2, G.c7, G.c8, H.c2, I.c3, J.c9).

## Correlation way to approach 

```python
ndf = df.iloc[:,2:]
for i in range(len(ndf)):
    if (ndf['class'][i]  == 'typ'):
        ndf['class'][i] = 1
    else:
        ndf['class'][i] = 0
ndf.head()
```

```python
ndf['class'] = ndf['class'].astype(str).astype(float)
```

```python
ndf.corr()
```

I select the first  line of correlation, and pick the value greater than 0.9 with 'class'.

```python
re = ndf.corr().iloc[0]
```

```python
for i in range(1, len(re)) :
    if (abs(re[i]) > .9):
        print(re.index[i])
```

According to the output, the significant channel are (A.c9, B.c1, B.c3, C.c1, E.c2, F.c2, G.c7, G.c8, H.c2, I.c3, J.c9).
